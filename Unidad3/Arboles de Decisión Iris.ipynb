```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
iris
```

```python
clf = tree.DecisionTreeClassifier()
##ver parametro son los valores de las medidas de las flores 
## el segundo parametro es la varible que se quiere predecir 
clf = clf.fit(iris.data,iris.target)

clf.score(iris.data,iris.target)

##genera la salida en formato .dot
## y lo podemos ver del archivo 2 con la libreria grap
tree.plot_tree(clf)
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph= graphviz.Source(dot_data)
graph.render('iris',format = 'png')
```

```python
dot_data = tree.export_graphviz(clf,out_file=None, feature_names=iris.feature_names,class_names=iris.target_names,filled=True,rounded=True,special_characters=True)
graph= graphviz.Source(dot_data)
graph.render("irisColor",format = "png")
graph
```
