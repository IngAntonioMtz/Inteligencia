```python
import pandas as pd
coronavirus = pd.read_csv("200723COVID19MEXICO.csv", encoding= 'unicode_escape')
coronavirus
```

```python
coronavirus.info()
```

```python
coronavirus['VIVIO'] = coronavirus['FECHA_DEF'].apply(lambda toLabel: 0 if toLabel =='9999-99-99' else 1)
y_target = coronavirus["VIVIO"].values
coronavirus.head(10)
```

```python
columns = ["UCI","TABAQUISMO","RENAL_CRONICA","OBESIDAD","SEXO", "EDAD", "EMBARAZO", "INTUBADO" ,"NEUMONIA", "DIABETES","EPOC","ASMA","INMUSUPR","HIPERTENSION","CARDIOVASCULAR" ]
X_input = coronavirus[list(columns)].values
```

```python
from sklearn import tree
Train = tree.DecisionTreeClassifier(criterion="entropy", max_depth=5)
Train = Train.fit(X_input, y_target)
Train.score(X_input,y_target)
```

```python
clasificacion=['No Vive', 'Vive']
```

```python
import sklearn
with open("covid.dot", 'w') as f:
  f = tree.export_graphviz(Train, out_file=f, feature_names=columns, class_names=clasificacion)
```
